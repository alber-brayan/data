```python
import matplotlib.pyplot as plt
# Graficar un gráfico de dispersión usando Pandas
df.plot.scatter(x='Columna_A', y='Columna_C')
plt.xlabel('Etiqueta eje X')
plt.ylabel('Etiqueta eje Y')
plt.title('Gráfico de dispersión')
plt.show()
```

```python
# Graficar un gráfico de dispersión usando Seaborn
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.scatterplot(data=df, x='Columna_C', y='Columna_B')
plt.xlabel('Etiqueta eje X')
plt.ylabel('Etiqueta eje Y')
plt.title('Gráfico de dispersión')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Crear la figura 3D y el eje
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar un gráfico de dispersión 3D usando Matplotlib
ax.scatter(df['Columna_A'], df['Columna_B'], df['Columna_C'], s=50, c=np.random.rand(420,3), edgecolor='black', alpha=0.2)

# Personalizar etiquetas y título
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Gráfico de dispersión 3D')

# Mostrar la gráfica
plt.show()
```
